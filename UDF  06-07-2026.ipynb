```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
def upper_name(name):
    return name.upper()

upper_udf=udf(upper_name,StringType())
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("JoinPractice").getOrCreate()

# DataFrame 1
data1 = [
    (101, "Alice",   "HR",      50000, "Mumbai",25),
    (102, "Bob",     "IT",      60000, "Pune",24 ),
    (103, "Charlie", "Finance", 70000, "Delhi",17),
    (104, "David",   "IT",      65000, "Chennai",22)
]

columns1 = ["emp_id", "emp_name", "department", "salary", "city",'age']
df1 = spark.createDataFrame(data1, columns1)
```

```python
df2 = df1.withColumn('emp_name', upper_udf(df1.emp_name))
df2.show()
```

```python
df2 = df1.withColumn('city' , upper_udf(df1.city))
df2.show()
```

```python
def lower_name(name):
    return name.lower()

lower_udf = udf(lower_name , StringType())
```

```python
df2= df1.withColumn('emp_name' , lower_udf(df1.emp_name))
df2.show()
```

```python
df2 = df1.withColumn('city' , lower_udf(df1.city))
df2.show()
```

```python
def bonus_salary(salary):
    return salary + 5000

bonus_udf = udf(bonus_salary , LongType())
```

```python
df2 = df1.withColumn('bonus' , bonus_udf(df1.salary))
df2.show()
```

```python
df1.show()
```

```python
def age_group(age):
    if age <= 17:
        return "Young"
    
    elif age < 35:
        return "Adult"
    else:
        return "Senior"
    
age_group_udf = udf(age_group , StringType())
df2 = df1.withColumn('age_group' , age_group_udf(df1.age))
df2.show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("JoinPractice").getOrCreate()

# DataFrame 1
data1 = [
    (101, "Alice",   "HR",      50000, "Mumbai",25,121312324),
    (102, "Bob",     "IT",      60000, "Pune",24,121312324),
    (103, "Charlie", "Finance", 70000, "Delhi",17,12131232),
    (104, "David",   "IT",      65000, "Chennai",18,121312324)
]

columns1 = ["emp_id", "emp_name", "department", "salary", "city",'age','aadhar']

aadhar1 = spark.createDataFrame(data1, columns1)
```

```python
def mask_aadhar(aadhar):
    masked_aadhar = str(aadhar)[:4] + "XXXXXX" + str(aadhar)[-4:]
    return masked_aadhar

mask_aadhar_udf = udf(mask_aadhar, StringType())

aadhar1 = aadhar1.withColumn("masked_aadhar", mask_aadhar_udf(aadhar1.aadhar))
aadhar1.show()
```
