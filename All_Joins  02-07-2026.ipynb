# DataFrame 1

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("JoinPractice").getOrCreate()

# DataFrame 1
data1 = [
    (101, "Alice",   "HR",      50000, "Mumbai"),
    (102, "Bob",     "IT",      60000, "Pune"),
    (103, "Charlie", "Finance", 70000, "Delhi"),
    (104, "David",   "IT",      65000, "Chennai"),
    (105, "Eva",     "HR",      55000, "Mumbai"),
    (106, "Frank",   "Sales",   52000, "Bangalore"),
    (107, "Grace",   "Finance", 75000, "Hyderabad"),
    (108, "Henry",   "IT",      62000, "Pune"),
    (109, "Ivy",     "Sales",   58000, "Delhi"),
    (110, "Jack",    "HR",      53000, "Mumbai")
]

columns1 = ["emp_id", "emp_name", "department", "salary", "city"]

df1 = spark.createDataFrame(data1, columns1)

df1.show()
```

#DataFrame 2

```python
data2 = [
    (101, "Laptop",     "Active",   2, "Mumbai"),
    (102, "Desktop",    "Active",   1, "Pune"),
    (103, "Monitor",    "Inactive", 2, "Delhi"),
    (104, "Laptop",     "Active",   1, "Chennai"),
    (105, "Printer",    "Active",   1, "Mumbai"),
    (111, "Scanner",    "Active",   2, "Kolkata"),
    (112, "Tablet",     "Inactive", 1, "Jaipur"),
    (113, "Mobile",     "Active",   3, "Pune"),
    (114, "Laptop",     "Active",   1, "Nagpur"),
    (115, "Desktop",    "Inactive", 2, "Delhi")
]

columns2 = ["emp_id", "asset_name", "asset_status", "asset_count", "asset_city"]

df2 = spark.createDataFrame(data2, columns2)

df2.show()
```

```python
from pyspark.sql.functions import *
```

# inner join 
#returns only matching records

```python
df1.join(df2 , df1.emp_id == df2.emp_id, "inner").show()
```

# left join


```python
df1.join(df2 , df1.emp_id == df2.emp_id , "left").show()
```

# right join

```python
df1.join(df2 , df1.emp_id == df2.emp_id , "right").show()
```

#full outer join 

```python
df1.join(df2 , df1.emp_id == df2.emp_id , "outer").show()
```

# left semi
returns only matching rows from left dataframe 

```python
df1.join(df2 , df1.emp_id == df2.emp_id , "left_semi").show()
```

```python
df2.join(df1 , "emp_id" , "left_semi").show()
```

# Left anti join 
returns rows from left that have NO match 

```python
df1.join(df2 , "emp_id" , "left_anti").show()
```

```python
df2.join(df1 , "emp_id" , "left_anti").show()
```

#cross join

```python
df1.crossJoin(df2).show()
```

# Broadcast join

```python
df1.join(broadcast(df2),df1.emp_id == df2.emp_id , "inner").show()
```
