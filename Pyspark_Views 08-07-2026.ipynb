```python
%sql
select * ,
row_number() over (order by salary desc) rn
from employee_info_df
```

```python
%sql 
select * ,
row_number() over (order by emp_id desc) rn
from employee_info_df
```

```python
%sql
select * ,
dense_rank() over (order by salary desc) rn
from employee_info_df
```

```python
%sql 
select * , 
ntile() over (order by salary desc) rn
from employee_info_df
```

```python
%sql
select * ,
lead(salary) over (order by salary desc) next_salary
from employee_info_df
```

```python
%sql
select full_name, salary ,
lead(salary) over (order by salary desc) next_salary
from employee_info_df
```

```python
%sql
select full_name, salary ,
lag(salary) over (order by salary desc) next_salary
from employee_info_df
```

```python
%sql
select full_name, 
first_value(salary) over (order by salary desc) highest_salary
from employee_info_df
```

```python
%sql
select full_name, 
last_value(salary) over (order by salary desc) highest_salary
from employee_info_df
```

```python
%sql
select full_name, salary,
last_value(salary) over (order by salary  rows between unbounded preceding and unbounded following) last_salary
from employee_info_df
```
